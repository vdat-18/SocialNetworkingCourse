# Tổng quan về GNN (Graph Neural Network)

## 1. Khái niệm GNN
**GNN (Graph Neural Network)** là một lớp mô hình học sâu được thiết kế để hoạt động trên dữ liệu có cấu trúc **đồ thị** (Graph).  
Trong dữ liệu đồ thị, các đối tượng (**node**) và mối quan hệ giữa chúng (**edge**) được biểu diễn dưới dạng đồ thị \( G = (V, E) \):  
- \( V \): Tập hợp các **đỉnh (nodes)**.  
- \( E \): Tập hợp các **cạnh (edges)** kết nối giữa các đỉnh.

**Mục tiêu**: GNN giúp học biểu diễn (**representation learning**) cho các node, cạnh hoặc toàn bộ đồ thị dựa trên cả đặc trưng của node và cấu trúc đồ thị.

---

## 2. Lý do sử dụng GNN
- **Dữ liệu dạng đồ thị phổ biến**: Xuất hiện trong các ứng dụng như:
  - Mạng xã hội
  - Mạng giao thông
  - Mạng protein
  - Hệ thống khuyến nghị

- **Học thông tin từ cấu trúc đồ thị**:  
   GNN tận dụng thông tin từ các node lân cận và cấu trúc liên kết để học đặc trưng hiệu quả.

- **Khả năng mở rộng**:  
   GNN mở rộng các mô hình học sâu thông thường (CNN, RNN) để hoạt động trên dữ liệu không có cấu trúc cố định.

---

## 3. Các thành phần chính của GNN
GNN bao gồm 3 thành phần cơ bản:

### **1. Node Features**:
Mỗi node $v \in V$ được gán một vector đặc trưng $h_v$.

### **2. Message Passing (Lan truyền tin)**:
Trong quá trình lan truyền tin, mỗi node thu thập thông tin từ các node lân cận thông qua các cạnh kết nối.

Công thức tổng quát:
$$
h_v^{(k)} = \text{AGGREGATE} \left( h_u^{(k-1)}, \forall u \in \mathcal{N}(v) \right)
$$

- $h_v^{(k)}$: Đặc trưng của node $v$ tại lớp thứ $k$.  
- $\mathcal{N}(v)$: Tập lân cận của node $v$.  
- **AGGREGATE**: Hàm tổng hợp thông tin (ví dụ: trung bình, tổng hoặc hàm phi tuyến).

### **3. Update**:
Sau khi thu thập thông tin từ lân cận, đặc trưng node được cập nhật:
$$
h_v^{(k)} = \text{UPDATE} \left( h_v^{(k-1)}, \text{AGGREGATE} \right)
$$

- **UPDATE**: Thường là một hàm phi tuyến như MLP hoặc ReLU.

---

## 4. Các loại GNN phổ biến
| **Mô hình**                     | **Đặc điểm**                                                                 |
|---------------------------------|-----------------------------------------------------------------------------|
| **GCN (Graph Convolutional Networks)** | Tích chập trên đồ thị để học đặc trưng node.                                  |
| **GraphSAGE**                   | Lấy mẫu node lân cận thay vì sử dụng toàn bộ lân cận.                        |
| **GAT (Graph Attention Networks)**    | Sử dụng cơ chế attention để học trọng số khác nhau cho các node lân cận.      |
| **GIN (Graph Isomorphism Network)**   | Mô hình mạnh mẽ giúp phân biệt các đồ thị khác nhau nhờ khả năng tổng hợp mạnh. |
| **R-GCN (Relational GCN)**            | GCN mở rộng cho đồ thị với các loại cạnh khác nhau (đồ thị quan hệ).          |

---

## 5. Ứng dụng của GNN
GNN có nhiều ứng dụng thực tế trong các lĩnh vực khác nhau:

### **1. Mạng xã hội**:
- Phân loại người dùng (**node classification**).  
- Dự đoán kết nối mới (**link prediction**).

### **2. Hệ thống khuyến nghị**:
- Khuyến nghị sản phẩm dựa trên mối quan hệ người dùng và sản phẩm.

### **3. Hóa học và sinh học**:
- Dự đoán tính chất của phân tử (**graph classification**).  
- Phân tích mạng protein tương tác (PPI).

### **4. Xử lý ngôn ngữ tự nhiên (NLP)**:
- Trích xuất quan hệ trong văn bản dựa trên **đồ thị tri thức**.

### **5. Mạng giao thông**:
- Dự đoán lưu lượng giao thông dựa trên mối liên hệ giữa các nút giao thông.

---

## 6. Ví dụ tổng quát về GNN
Quy trình học trên GNN cơ bản có thể mô tả như sau:
1. Mỗi node được khởi tạo với đặc trưng ban đầu $h_v^{(0)}$.
2. Tại mỗi lớp $k$:
   - Node thu thập thông tin từ lân cận qua hàm **AGGREGATE**.
   - Node cập nhật đặc trưng của mình qua hàm **UPDATE**.
3. Sau $K$ lớp lan truyền tin, ta nhận được đặc trưng $h_v^{(K)}$ cho node $v$.

**Đặc trưng này có thể được sử dụng để**:
- Phân loại node (**node classification**).  
- Dự đoán liên kết (**link prediction**).  
- Phân loại đồ thị (**graph classification**).

---

## 7. Ưu điểm và hạn chế của GNN

### **Ưu điểm**:
- Học được thông tin từ cả đặc trưng của node và cấu trúc đồ thị.
- Khả năng mở rộng với các loại dữ liệu phi cấu trúc.

### **Hạn chế**:
- Hiệu quả tính toán thấp trên đồ thị lớn do quá trình lan truyền tin.
- Có thể gặp vấn đề **over-smoothing** khi số lớp GNN quá lớn (node có đặc trưng giống nhau).

---

## Tổng kết
GNN là một công cụ mạnh mẽ giúp học trên dữ liệu dạng đồ thị, với nhiều biến thể và ứng dụng trong thực tế như **mạng xã hội**, **hệ thống khuyến nghị**, và **phân tích sinh học**.  
Các mô hình như **GCN**, **GAT**, và **GraphSAGE** giúp GNN có khả năng mở rộng và xử lý dữ liệu đồ thị phức tạp.


```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
class LinkPrediction:
    def __init__(self, filepath):
        """
        Khởi tạo với file danh sách cạnh đã cho.

        Parameters:
            filepath (str): Đường dẫn tới file danh sách cạnh.
        """
        print("\nĐang đọc file danh sách cạnh...")
        self.G = nx.read_edgelist(filepath, create_using=nx.DiGraph(), nodetype=int)
        print("Đọc file thành công!")

    def analyze_network(self):
        """Phân tích cơ bản về mạng."""
        print("\n=== Phân tích mạng ===")
        print(f"Số node: {self.G.number_of_nodes()}")
        print(f"Số cạnh: {self.G.number_of_edges()}")
        print(f"Clustering trung bình: {nx.average_clustering(self.G.to_undirected()):.3f}")

        plt.figure(figsize=(15, 5))
        pos = nx.spring_layout(self.G, seed=42)
        nx.draw(self.G, pos, with_labels=True, node_color='lightblue', node_size=500)
        plt.title("Cấu trúc mạng")
        plt.show()

    def prepare_data(self):
        """Chuẩn bị dữ liệu train và test."""
        edges = list(self.G.edges())
        np.random.seed(42)
        np.random.shuffle(edges)

        n_test = int(len(edges) * 0.2)
        self.test_edges = edges[:n_test]
        self.train_edges = edges[n_test:]

        self.train_G = self.G.copy()
        self.train_G.remove_edges_from(self.test_edges)

        non_edges = list(nx.non_edges(self.train_G))
        np.random.shuffle(non_edges)
        self.test_non_edges = non_edges[:n_test]
        print(f"Train edges: {len(self.train_edges)}")
        print(f"Test positive edges: {len(self.test_edges)}")
        print(f"Test negative edges: {len(self.test_non_edges)}")

    def extract_features(self, G, u, v):
        """Trích xuất features cho các node pairs."""
        undirected_G = G.to_undirected()  # Chuyển thành đồ thị vô hướng tạm thời
        cn = len(list(nx.common_neighbors(undirected_G, u, v)))
        jc = list(nx.jaccard_coefficient(undirected_G, [(u, v)]))[0][2]
        aa = list(nx.adamic_adar_index(undirected_G, [(u, v)]))[0][2]
        pa = list(nx.preferential_attachment(undirected_G, [(u, v)]))[0][2]
        return [cn, jc, aa, pa]

    def prepare_ml_data(self, G, positive_edges, negative_edges):
        """Tạo dữ liệu cho machine learning."""
        undirected_G = G.to_undirected()  # Chuyển toàn bộ đồ thị thành vô hướng
        X = [self.extract_features(undirected_G, u, v) for u, v in positive_edges + negative_edges]
        y = [1] * len(positive_edges) + [0] * len(negative_edges)
        return np.array(X), np.array(y)

    def evaluate_baselines(self):
        """Đánh giá các phương pháp truyền thống."""
        # Chuyển đồ thị có hướng thành vô hướng
        undirected_train_G = self.train_G.to_undirected()

        methods = {
            'Common Neighbors': lambda u, v: len(list(nx.common_neighbors(undirected_train_G, u, v))),
            'Jaccard Coefficient': lambda u, v: list(nx.jaccard_coefficient(undirected_train_G, [(u, v)]))[0][2],
            'Adamic-Adar': lambda u, v: list(nx.adamic_adar_index(undirected_train_G, [(u, v)]))[0][2],
            'Preferential Attachment': lambda u, v: list(nx.preferential_attachment(undirected_train_G, [(u, v)]))[0][2],
        }

        results = []

        for name, score_func in methods.items():
            # Tính điểm cho positive và negative edges
            pos_scores = [score_func(u, v) for u, v in self.test_edges]
            neg_scores = [score_func(u, v) for u, v in self.test_non_edges]

            # Tạo y_true và y_scores
            y_true = [1] * len(pos_scores) + [0] * len(neg_scores)
            y_scores = pos_scores + neg_scores

            # Tính AUC
            auc = roc_auc_score(y_true, y_scores)

            # Lưu kết quả
            results.append({
                "Method": name,
                "AUC": auc
            })

        # Chuyển kết quả thành DataFrame
        self.results_df = pd.DataFrame(results)
        print("\nBảng kết quả so sánh các phương pháp:")
        print(self.results_df)

    def evaluate_ml(self):
        """Huấn luyện mô hình Machine Learning và đánh giá."""
        X, y = self.prepare_ml_data(self.train_G, self.test_edges, self.test_non_edges)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        model = RandomForestClassifier(random_state=42)
        model.fit(X_train, y_train)
        y_pred_proba = model.predict_proba(X_test)[:, 1]

        auc = roc_auc_score(y_test, y_pred_proba)
        print(f"\nAUC của Random Forest: {auc:.3f}")

    def run_experiment(self):
        """Chạy thử nghiệm."""
        self.analyze_network()
        self.prepare_data()
        self.evaluate_baselines()
        self.evaluate_ml()
```
